# inputing catagorical data with mode value

```python
for value in ['Gender','Married','Dependents','Self_Employed','Loan_Amount_Term']:
    risk_analytics_train[value].fillna(risk_analytics_train[value].mode()[0],inplace=True)
```

```python
print(risk_analytics_train.isnull().sum())
```

# imputing numeriacal missing values with mean value  
 note- 0 means fill values for columns , 1 means fill values for rows in code

```python
risk_analytics_train["LoanAmount"].fillna(round(risk_analytics_train["LoanAmount"].mean(),0),inplace=True)
```

```python
print(risk_analytics_train.isnull().sum())
```

# imputing values for credit history column differently

```python
risk_analytics_train["Credit_History"].fillna(value=0,inplace=True)
risk_analytics_train.Credit_History.mode()
```
